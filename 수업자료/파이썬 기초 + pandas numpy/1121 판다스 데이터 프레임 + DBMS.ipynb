```python
# database 에서 sql로 데이터를 추출해서 pandas의 DataFrame으로 생성 

## 사용하는 DBMS는 MySQL , 사용하는 데이터는 book table
# 지금은 강사님 컴에 접속 
# pymysql 설치하세요 

import pymysql.cursors # mysql DBMS를 사용하기 위한 모듈을 로딩 
import numpy as np 
import pandas as pd 

# 데이터베이스 커넥션 
con= pymysql.connect(host = "70.12.116.160",user="data",password = "data",db="library",charset="UTF8")

keyword = "여행"
# DBMS에서 data를 가져오기 위한 SQL 을 작성
# 책 제목에 키워드가 들어있는 행을 찾아서 제목 가격 저자정보를 가져와  # \ 는 줄바꿈 기호
sql = "select btitle,bprice,bauthor "+\
      "from book " +\
      "where btitle like '%{}%'".format(keyword)

try:
    df= pd.read_sql(sql,con)
    # 정상적으로 데이터프레임을 얻어왔으면 
    # 제이슨 파일로 변환하세요 
    df.to_json("./data/book_columns.json",orient = "columns")# 컬럼을 기반으로 제이슨 만들거야
    df.to_json("./data/book_records.json",orient = "records")
    df.to_json("./data/book_index.json",orient = "index")
    df.to_json("./data/book_values.json",orient = "values")
except Exception as inst:
    print("뭔가 틀렷다")
finally:
    con.close()   # 처리가 되던 안되던 닫아
```
