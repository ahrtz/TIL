```python
# 상관계수 상관관계  (CoRelation)

# 두개의 확률 변수가 어떤 관계인지 보여줌 

# 두대상이 서로 연관성이 있다고 추측되는 관계 :
# 일반적으로 피어슨 상관계수를 이용한다 다른 상관계수도 존재 
"상관계수 = correlation coefficient"
# 공분산을 각 확률변수의 표준편차의 곱으로 나눠준값  (-1~1)
# 양의 값이면 양의 상관관계 
#음의 값이면 부적상관관계
# 상관계수는 인과관계를 설명하지 않는다. 
# 단지 관련성이 있다는 것만 표현해줌 
# 인과관계는 회귀분석(regression ) 을 통해서 알 수 있다. 
```

```python
import numpy as np
import pandas as pd 
import json

file_KOSPI = open("C:/python_DA/data/stock/KOSPI.json","r")
df_KOSPI = pd.DataFrame(json.load(file_KOSPI))
series_KOSPI= df_KOSPI["Close"]
file_SE = open("C:/python_DA/data/stock/SE.json","r")
df_SE = pd.DataFrame(json.load(file_SE))
series_SE= df_SE["Close"]

file_PU = open("C:/python_DA/data/stock/부산산업.json","r")
df_PU = pd.DataFrame(json.load(file_PU))
series_PU= df_PU["Close"]

## 상관계수를 구하기 위해서 일단 공분산을 구한후 각 표준편차의 곱으로 나눠 줘야 한다.
" 주의 np가 가지고 있는 함수를 사용한다. CORelation corffincent -> np.corrcoef()"
#print(np.corrcoef(series_KOSPI,series_SE)) # 0.9 는 거의 비슷하다 한개가 올라가면 다른 한개도 거의 올라간다 
# 선후 관계 모름, 두개만 됨
data = {"KOSPI" : series_KOSPI,
       "SE" : series_SE,
       "PU" : series_PU}
df = pd.DataFrame(data)
df.corr()
df.cov()
```

```python
#중첩리스트를 사용해 데이터 프레임 만들기 
data = [[2,np.nan],
       [7,-3],
       [np.nan,np.nan],
       [1,-2]]
df=pd.DataFrame(data,columns=["one","two"],index=["a","b","c","d"])
df.sum() # axis를 명시하지 않으면 =0 으로 간주 열 방향 세팅 # series 로 반환 NaN 0으로 간주 
df.mean() # nan 자동으로 빼고 계산한다 
df["two"].sum()
df.loc["b"].sum()
df.mean(skipna=False) # nan 실수처리 모든 연산결과는 NAN ,Default 값은 트루 

"가지고 놀기"
# one column 의 결측값은 one 컬럼의 평균
# 투 컬럼의 결측값은 투 컬럽의 최소값으로 대체
df["one"][df["one"].isna()]=df["one"].mean()
df["two"][df["two"].isna()]=df["two"].min()
#df["one"].replace(np.nan,df["one"].mean())
#df["two"].replace(np.nan,df["two"].min())

# 강사님 예시
# df["one"]=df["one"].fillna(value=df["one"].mean())
# df["two"]=df["two"].fillna(value=df["two"].min())
```

```python
# unique( ), isin(), value_count( ) 함수
import numpy as np 
import pandas as pd 

np.random.seed (1)
data= np.random.randint(0,10,(6,4))
df = pd.DataFrame(data)
df.columns= ["A","B","C","D"]
df.index = pd.date_range("20190101",periods = 6)
# 새로운 컬럼 만들기
df["E"]= ["AA","BB","CC","BB","AA","AA"]
df["E"].unique() # ndarray로 리턴해준다 set이라고 생각하면 된다. 
df["E"].value_counts() # 몇번 들어가있는지 
print(df["E"].isin(["AA","CC"])) # 불리언 마스트를 리턴한다. 
display(df)
```

```python
data1= {
    "학번":[1,2,3,4],
    "이름" : ["홍길동","임꺽정","이순신","강감찬"],
    "학년" : [2,4,4,3]
}
data2 = {
    
    "학과": ["컴퓨","미술","철학","사회"],
    "학점": [1.3,3.5,4.3,2.3]
}
df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
df2.index=[1,2,4,5]

pd.merge(df1,df2,left_on="학번",right_index=True,how="inner")   # 컬럼이 아니라 인덱스 번호와 연결 시키고 싶을때
```
