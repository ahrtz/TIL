```python
# 데이터 프레임의 다중인덱스

np.random.seed(100)


df= pd.DataFrame(np.random.randint(1,10,(6,5)),index=[["총무처","관리처","정보전산처","총무처","관리처","정보전산처"],
                                                ["인력개발팀","건축팀","개발팀","재무팀","안전팀","운영팀"]],
                columns=[["아시아","유럽","아시아","아시아","유럽"],["한","프","일","중","독"]])
display(df)
# 정렬해보자
new_df = df.sort_index(axis=1) # 열 방향 정렬 ( 컬럼 정렬)
new_df = new_df.sort_index(axis=0) # 행방향 정렬 (row 정렬)
new_df

df[("아시아","한")]
df.loc["총무처"] # 행 인덱스는 loc사용
df.loc["총무처","재무팀"]

# 아시아 지역의 중국의 정보전산처 개발팀인원은? 
df["아시아","중"].loc["정보전산처","개발팀"]
df.loc[("정보전산처","개발팀"),("아시아","중")]
```

```python

np.random.seed(1000)


df= pd.DataFrame(np.random.randint(1,10,(6,5)),index=[["총무처","관리처","정보전산처","총무처","관리처","정보전산처"],
                                                ["인력개발팀","건축팀","개발팀","재무팀","안전팀","운영팀"]],
                columns=[["아시아","유럽","아시아","아시아","유럽"],["한","프","일","중","독"]])


df.index.names = ["부서","팀"]
df.columns.names = ["대륙","국가"]
display(df)

#팀이름으로 내림차순 정렬
#df.sort_index(axis=0,by="팀",ascending=False)      <- 되긴하는데 경고 뜬다
df.sort_index(axis=0,level="팀",ascending=False)
df.sort_index(axis=0,level=1,ascending=False)      # 둘다 된다 
df.sort_index(axis=1,level=1,ascending=True)
df.sort_index(axis=1,level=0,ascending=True)

# 값을 이용해서 정렬하기 
df.sort_values(axis=0,by=("아시아","한"),ascending=False)   ## 아시아의 한국을 기준으로 정렬 내림차순정렬 
                                                            ## by 뒤에는 튜플을 사용해 주어야 한다!! 리스트 안댐
# 아시아 지역의 한국직원수를 기준으로 내림차순 정렬하고 2차정렬에 대해서는 아시아 지역의 일본직원수를 기준으로 오름차순 정렬
"두개를 기준으로 할때 (1차 2차 소팅)"
df.sort_values(axis=0,by=[("아시아","한"),("아시아","일")],ascending=[False,True])  
```

```python
# 데이터 프레임에 대한 sum() mean()

np.random.seed(1000)


df= pd.DataFrame(np.random.randint(1,10,(6,5)),index=[["총무처","관리처","정보전산처","총무처","관리처","정보전산처"],
                                                ["인력개발팀","건축팀","개발팀","재무팀","안전팀","운영팀"]],
                columns=[["아시아","유럽","아시아","아시아","유럽"],["한","프","일","중","독"]])


df.index.names = ["부서","팀"]
df.columns.names = ["대륙","국가"]
display(df)

# 각 도시별 직원수의 합은?
df.sum(axis=0)
#각 도시 부처별 직원수의 합은?
df.sum(axis=0,level="부서")
df.sum(axis=0,level=1)
##모든지역의 각 부서 팀별 평균인원
df.mean(axis=1)
```

```python
import numpy as np
import pandas as pd
np.random.seed(1)
df = pd.DataFrame(np.random.randint(1,5,(2,3)),index = ["서울","부산"],
                 columns=["마케팅","개발","운영"])
new_df=df.stack()
display(new_df)  # 컬럼을 인덱스로 바꾸는것입니다. 
# 기존인덱스의 하위로 밀어버림 
# 이거의 반대가 unstack

origin_df=new_df.unstack(level=0) # level 이 0 이면 제일 큰 범주를 뺄수 있다 대단위 --> 레벨 설정으로 가능할듯 
                                  #3중구조일때 되는지는 실험해봐야 할듯

origin_df
```

```python
# 중복행 제거 (추출)

df= pd.DataFrame({"k1":["one"]*3+["two"]*4,
                 "k2":[1,1,2,3,3,4,4]})
display(df)
#중복행 쉽게 찾기
df.duplicated() # boolean mask 
df.loc[df.duplicated()]    # 중복항 뽑기
df.loc[~df.duplicated()]       # 아닌 거 뽑기 

df.drop_duplicates()
## 중복항 제거 하기 
```

```python
# DataFrame 의 grouping
import numpy as np
import pandas as pd
df = pd.DataFrame({"이름":["홍길동","김연아","아이유","김길동","신사임당"],
                  "성적": [1.5,4.4,4.5,2.3,4.0],
                  "학과":["컴퓨터","철학","컴퓨터","철학","컴퓨터"],
                  "학년":[1,2,3,2,2]})
display(df)

# 학과를 기준으로 그룹핑
dept=df["성적"].groupby(df["학과"])
dept.mean()
# 학과와 학년을 기준으로 성적을 grouping
dept=df["성적"].groupby([df["학과"],df["학년"]])
print(dept.mean())
dept.mean().unstack()
```
