```python
## 데이터를 그래프로 그려보기 
## conda install matplotlib
import numpy as np
import matplotlib.pyplot as plt

arr = np.linspace(0,10,20)
plt.plot(arr)
plt.show()
```

```python
## numPy array(ndarray)

# 난수를 이용해서 ndarray생성하기 총 다섯가지 

# np.random.normal () : 난수를 생성해서 ndarray를 만들 수 있다.
# ==> 정규 분포 확률 밀도에서 난수를 발생 시킨다. 
# 평균과 표준편차 필요
myMean = 50 
myStd = 3
arr = np.random.normal(myMean,myStd,(10000,2)) # 표준 편차 뒤에 shape 을 넣어준다
plt.hist(arr,bins=400)# bins는 구간을 의미
#plt.show
#arr = np.random.random((2,4))  
print(arr)
```

```python
# 난수를 추출 
#[0,1) 구간에서 균등분포 확률밀도상에서 난수 추출
arr = np.random.rand(1000,3) # 3차원 10개의면 2개의 행 3개의 열 # 인자를 2개만 주면 행 렬 1개면 행
#print(arr)
plt.hist(arr,bins=300)
plt.show()
```

```python
# 난수를 추출 
# 표준 정규 분포에서 난수 추출  (정규분포를 평균 0 ,sd =1 로 변환)
arr=np.random.randn(10000,1)  # 면 행 열  # 인자의 개수로 차원을 지정
#print(arr)
plt.hist(arr,bins = 300)
plt.show()
```

```python
## 난수를 추출(정수를 추출)
# 주어진 범위에서 균등분포 확률밀도상에서 난수 추출 
arr= np.random.randint(-100,100,(10000000,))     # 범위 시작,끝,shape
print(arr)
plt.hist(arr,bins = 300)
plt.show()
```

```python
# 난수를 추출(실수)
# [0,1) 균등분포로 실수 추출
arr = np.random.random((100000,)) # 3차원 10개의면 2개의 행 3개의 열 # 인자를 2개만 주면 행 렬 1개면 행
# 얘는 shape 을 줘야한다 .rand랑 비교 
plt.hist(arr,bins=300)
plt.show()
```

```python
## 난수는 랜덤값이라 실행할때마다 추출되는 난수가 달라져야 한다. 
## 난수값도 사실은 알고리즘에 의해 계산되는값임 
# 알고리즘의 초기값을 고정시키면 항상 같은 난수를 얻을수 있다. => 재현성

np.random.seed(30)

arr = np.random.randint(0,10,(5,)) 
print(arr)
```
